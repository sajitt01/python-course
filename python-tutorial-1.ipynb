### List

So far we've been using single variables and setting them to a, b, c,
etc. But what about if we have a how set of data. Say a time series of
values, we want to work with the whole set of data at once. For this we
can use special variables like `lists` and `dictionaries`

Below is some code to make a `list` full of numbers. Lists are very
general, and can contain basically anything, even other lists. What do
you think the output of the following code will be?

```
a = [1, 2, 3, 4, 5]
b = ['abce', 10, 1.0, '3', a]
print(a)
print(b)
```

Now, what do you think will happen if we say? What will c be?

```
a = [1, 2, 3, 4]
b = ['10', 'nothing']
c = a + b
print(c)
```

And then from that, what about the following? What will c be?

```
a = [1, 2, 3]
b = 10
c = a + b
print(c)
```

We can get a value in a list by using its <u>index</u> - this is the
position in the list. What will happen if we run the following code?

```
a = [1, 2, 3]
print(a[1])
```

Python is <u>zero indexed</u> that means that it starts counting from
0 - this can be confusing at first but ends up being simpler for more
complex operations. Just remember, that if you want the second thing in
a list, you want to use `1` because its going to start from 0 not 1.

### Dictionaries

Dictionaries are like lists, but rather than having all the data
ordered, they use a `key` paired to each value in the dictionary. This
can be useful if your data isn't ordered, but you have specific value to
connect it to. For example, if we had 5 students in a class, who had all
done a test

```
student_names = ['anna', 'andrew', 'walter', 'suzy', 'bob']
scores = [85, 20, 50, 0, 99]
```

We could just remember that anna is the first in the scores list, but
what about if we had 1000 students? Its easier to be able to ask "what
was anna's score"? We will use a dictionary for this

```
student_scores = {'anna':85, "andrew":20, 'walter':50, 'suzy':0, 'bob':99}
print(student_scores)
```

Each piece of data in a dictionary is referred to by its `key`, rather
than its position. So `'anna':85` can be read as `key:value`. If we want
to get the score for a student, we use their name (or key). So what will
we get if we run:

```
print(student_scores['anna'])
```

And what about if we try the following. Will it run?

```
print(student_scores[1])
```
